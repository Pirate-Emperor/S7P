```python
data = pd.read_csv("PTweet_WWE.csv")
data.head()
```

```python
df = pd.DataFrame(data['text'])
df.head()
```

### 1. Lower Casing

```python
# Task 1: Lowercasing
df['lowercased_text'] = df['text'].apply(lambda x: x.lower())
df.head()
```

### 2. Tokenization

```python
# Task 2: Tokenization
df['tokens'] = df['lowercased_text'].apply(lambda x: re.findall(r'\b\w+\b', x))
# df['tokens'] = df['lowercased_text'].apply(lambda x: word_tokenize(x))
df.head()
```

### 3. Punctuation Mark Removal

```python
# Task 3: Punctuation Mark Removal
df['cleaned_text'] = df['tokens'].apply(lambda x: ''.join(char for char in x if char not in string.punctuation))
df.head()
```

### 4. Stop Word Removal

```python
# Task 4: Stop Word Removal
stop_words = set(stopwords.words('english'))
df['filtered_text'] = df['tokens'].apply(lambda x: ' '.join(word for word in x if word not in stop_words))
df.head()
```

### 5. Stemming

```python
# Task 5: Stemming
stemmer = PorterStemmer()
df['stemmed_Text'] = df['tokens'].apply(lambda x: ' '.join(stemmer.stem(word) for word in x))
df.head()
```

### 6. Lemmatization

```python
# Task 6: Lemmatization
lemmatizer = WordNetLemmatizer()
df['lemmatized_text'] = df['tokens'].apply(lambda x: ' '.join(lemmatizer.lemmatize(word) for word in x))
df.head()
```

```python
for index, row in df.iterrows():
  for token in row['tokens']:
    lemmatized_word = lemmatizer.lemmatize(token)
    print(f"Token: {token}, Lemmatized: {lemmatized_word}, Equal: {token == lemmatized_word}")
```

```python
equal_counts = []
unequal_counts = []

for index, row in df.iterrows():
  equal_count = 0
  unequal_count = 0
  for token in row['tokens']:
    lemmatized_word = lemmatizer.lemmatize(token)
    if token == lemmatized_word:
      equal_count += 1
    else:
      unequal_count += 1
  equal_counts.append(equal_count)
  unequal_counts.append(unequal_count)

df['simple_count'] = equal_counts
df['complex_count'] = unequal_counts

df.head(40)
```

### 7. Translation

```python
# Task 7: Translation
# translator = google_translator()
df['translated_text'] = df['lowercased_text'].apply(lambda x: GoogleTranslator(source='auto', target='es').translate(x))  # Translate to Spanish
df.head()
```

### 8. Emoji to text

```python
# Task 8: Emoji to Text
df['emoji_to_text'] = df['text'].apply(lambda x: emoji.demojize(x))
df.head()
```

```python
import re
import string

# Sample text data
text_data = data.head()['text']

# Task 1: Lowercasing
lowercased_texts = [text.lower() for text in text_data]

# Task 2: Tokenization
tokenized_texts = [re.findall(r'\b\w+\b', text) for text in text_data]

# Task 3: Punctuation Mark Removal
cleaned_texts = [''.join(char for char in text if char not in string.punctuation) for text in text_data]

# Task 4: Stop Word Removal
stop_words = set(["a", "an", "the", "is", "from", "this"])
filtered_texts = [' '.join(word for word in text.split() if word.lower() not in stop_words) for text in text_data]

# Task 5: Stemming
def simple_stemming(text):
    return ' '.join(word[:4] if len(word) > 4 else word for word in text.split())

stemmed_texts = [simple_stemming(text) for text in text_data]

# Task 6: Lemmatization
def simple_lemmatization(text):
    return ' '.join(word[:-2] if word.endswith("es") else word for word in text.split())

lemmatized_texts = [simple_lemmatization(text) for text in text_data]

# Display results
for i in range(len(text_data)):
    print(f"\nOriginal Text: {text_data[i]}")
    print(f"Lowercased Text: {lowercased_texts[i]}")
    print(f"Tokenized Text: {tokenized_texts[i]}")
    print(f"Cleaned Text: {cleaned_texts[i]}")
    print(f"Filtered Text: {filtered_texts[i]}")
    print(f"Stemmed Text: {stemmed_texts[i]}")
    print(f"Lemmatized Text: {lemmatized_texts[i]}")
```

```python
import pandas as pd
import nltk
from nltk.util import ngrams
from collections import Counter

def tokenize_text(text):
    return nltk.word_tokenize(text.lower())

def generate_ngrams(text, n):
    tokens = tokenize_text(text)
    n_grams = list(ngrams(tokens, n))
    return n_grams

def ngram_analysis(df, n):
    all_ngrams = []
    for text in df['filtered_text']:
        ngrams = generate_ngrams(text, n)
        all_ngrams.extend(ngrams)
    return Counter(all_ngrams)

def top_ngrams(ngram_counts, top_n):
    top_ngrams_list = ngram_counts.most_common(top_n)
    ngram_strings = [' '.join(gram) for gram, count in top_ngrams_list]
    counts = [count for gram, count in top_ngrams_list]
    return pd.DataFrame({'N-gram': ngram_strings, 'Count': counts})

unigram_counts = ngram_analysis(df, 1)
bigram_counts = ngram_analysis(df, 2)
trigram_counts = ngram_analysis(df, 3)

n_values = [10]

def display_top_ngrams():
    for n in n_values:
        print(f"Top {n} Uni-grams:")
        print(top_ngrams(unigram_counts, n), "\n")

        print(f"Top {n} Bi-grams:")
        print(top_ngrams(bigram_counts, n), "\n")

        print(f"Top {n} Tri-grams:")
        print(top_ngrams(trigram_counts, n), "\n")

display_top_ngrams()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd

classifier_data = {
    'text': new_data['text'],
    'label': new_data['popular']  # 1 = popular, 0 = not popular
}
df = pd.DataFrame(classifier_data)
print(f"Evaluating on WWE twitter dataset...")
print(df.head())
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['label'], test_size=0.3, random_state=42)
def evaluate_model(ngram_range, ngram_type):
    vectorizer = TfidfVectorizer(ngram_range=ngram_range)
    X_train_vec = vectorizer.fit_transform(X_train)
    X_test_vec = vectorizer.transform(X_test)
    clf = LogisticRegression()
    clf.fit(X_train_vec, y_train)
    y_pred = clf.predict(X_test_vec)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    return {
        'ngram_type': ngram_type,
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1_score': f1
    }
print("Evaluation Metrics: ")
results = []
results.append(evaluate_model((1, 1), 'Uni-grams'))
results.append(evaluate_model((1, 2), 'Uni + Bi-grams'))
results.append(evaluate_model((1, 3), 'Uni + Bi + Tri-grams'))
results.append(evaluate_model((2, 2), 'Bi-grams'))
results.append(evaluate_model((3, 3), 'Tri-grams'))
results.append(evaluate_model((2, 3), 'Bi + Tri-grams'))
results_df = pd.DataFrame(results)
print(results_df)
```

```python
import nltk
from nltk.corpus import reuters, twitter_samples, product_reviews_1, names, webtext
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd

nltk.download('reuters')
nltk.download('twitter_samples')
nltk.download('product_reviews_1')
nltk.download('names')
nltk.download('webtext')

def prepare_movie_reviews():
    documents = [(list(movie_reviews.words(fileid)), category)
                 for category in movie_reviews.categories()
                 for fileid in movie_reviews.fileids(category)]
    reviews = [' '.join(doc) for doc, _ in documents]
    labels = [1 if label == 'pos' else 0 for _, label in documents]
    return reviews, labels

def prepare_reuters():
    categories = reuters.categories()
    documents = [(list(reuters.words(fileid)), category)
                 for fileid in reuters.fileids()
                 for category in categories if category in reuters.categories(fileid)]
    reviews = [' '.join(doc) for doc, _ in documents]
    labels = [categories.index(label) for _, label in documents]
    return reviews, labels

def prepare_twitter_samples():
    positive_tweets = twitter_samples.strings('positive_tweets.json')
    negative_tweets = twitter_samples.strings('negative_tweets.json')
    reviews = positive_tweets + negative_tweets
    labels = [1] * len(positive_tweets) + [0] * len(negative_tweets)
    return reviews, labels

def prepare_product_reviews():
    print(product_reviews_1.review())
    reviews = [review.text for review in product_reviews_1.reviews()]
    labels = [review.rating > 3 for review in product_reviews_1.reviews()]
    return reviews, labels

def prepare_names():
    male_names = names.words('male.txt')
    female_names = names.words('female.txt')
    reviews = male_names + female_names
    labels = [1] * len(male_names) + [0] * len(female_names)
    return reviews, labels

def prepare_webtext():
    reviews = [' '.join(webtext.words(fileid)) for fileid in webtext.fileids()]
    labels = [i for i in range(len(reviews))]
    return reviews, labels

datasets = [
    ('Movie Reviews', prepare_movie_reviews),
    # ('Reuters', prepare_reuters),
    ('Twitter Samples', prepare_twitter_samples),
    # ('Product Reviews', prepare_product_reviews),
    # ('Names', prepare_names),
    # ('Webtext', prepare_webtext)
]

def evaluate_model(ngram_range, ngram_type, X_train, X_test, y_train, y_test):
    vectorizer = TfidfVectorizer(ngram_range=ngram_range)
    X_train_vec = vectorizer.fit_transform(X_train)
    X_test_vec = vectorizer.transform(X_test)
    clf = LogisticRegression(max_iter=1000)
    clf.fit(X_train_vec, y_train)
    y_pred = clf.predict(X_test_vec)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    return {
        'ngram_type': ngram_type,
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1_score': f1
    }

for dataset_name, dataset_fn in datasets:
    print(f"Evaluating on {dataset_name} dataset...")
    reviews, labels = dataset_fn()
    dataset_df = pd.DataFrame({'text': reviews, 'label': labels})
    print(dataset_df.head())
    X_train, X_test, y_train, y_test = train_test_split(reviews, labels, test_size=0.3, random_state=42)
    results = []
    results.append(evaluate_model((1, 1), 'Uni-grams', X_train, X_test, y_train, y_test))
    results.append(evaluate_model((1, 2), 'Uni + Bi-grams', X_train, X_test, y_train, y_test))
    results.append(evaluate_model((1, 3), 'Uni + Bi + Tri-grams', X_train, X_test, y_train, y_test))
    results.append(evaluate_model((2, 2), 'Bi-grams', X_train, X_test, y_train, y_test))
    results.append(evaluate_model((3, 3), 'Tri-grams', X_train, X_test, y_train, y_test))
    results.append(evaluate_model((2, 3), 'Bi + Tri-grams', X_train, X_test, y_train, y_test))
    print("Evaluation Metrics: ")
    results_df = pd.DataFrame(results)
    print(results_df)
```

```python
from transformers import BertTokenizer, BertModel
import torch
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

def get_bert_embedding(ngram):
    inputs = tokenizer(ngram, return_tensors='pt')
    outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).detach().numpy()

def compute_similarity(embeddings):
    return cosine_similarity(embeddings)

def semantic_analysis(ngrams_list):
    embeddings = []
    for ngram in ngrams_list:
        embedding = get_bert_embedding(ngram)
        embeddings.append(embedding)
    embeddings = np.vstack(embeddings)
    similarity_matrix = compute_similarity(embeddings)
    return pd.DataFrame(similarity_matrix, index=ngrams_list, columns=ngrams_list)

n = 10
top_n_unigrams = top_ngrams(unigram_counts, n)['N-gram']
top_n_bigrams = top_ngrams(bigram_counts, n)['N-gram']
top_n_trigrams = top_ngrams(trigram_counts, n)['N-gram']

print("Semantic Analysis for Uni-grams:")
print(semantic_analysis(top_n_unigrams), "\n")

print("Semantic Analysis for Bi-grams:")
print(semantic_analysis(top_n_bigrams), "\n")

print("Semantic Analysis for Tri-grams:")
print(semantic_analysis(top_n_trigrams), "\n")
```
